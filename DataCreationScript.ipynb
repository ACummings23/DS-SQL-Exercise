```python
#####data set creation
import pandas as pd
from random import choice,seed
from string import ascii_lowercase, digits
```

```python
##### Business App to Server #####


### 3 tables linked using Applications


###Business Units table
BU_List_Desc=['fire safety',
'hydroelectric repairs',
'emergency storage',
'water damage',
'reactor protocols',
'dam controls',
'eletrical grid',
'coal storage',
'storm preparation',
'powergrid shutoff',
'auxilary power',
'solar energy controls and storage'
]
BU_List=['BU_1','BU_2','BU_3','BU_4','BU_5','BU_6','BU_7','BU_8','BU_9','BU_10','BU_11','BU_12']
BU=pd.DataFrame()
BU['BU']=BU_List
BU['BU Desc']=BU_List_Desc
BU.to_csv('BU_Desc.csv')
```

```python
##### create Applications table ####
ObjectState=['Active','Planned','Retired']
seed(20)

Asset_Type=[
    'Storage','Compute','Network','Software'
]
App_Name=[]
for i in range(100):
    App_Name.append('APP-'+str(i))
Line_of_Business=[
    'Power Generation',
    'Eletrical Grid',
    'Management',
    'Infrastructure',
    'IT'
]
Object_List=[''.join(choice(ObjectState) for _ in range(1)) for _ in range(100)]
Asset_List=[''.join(choice(Asset_Type) for _ in range(1)) for _ in range(100)]
LOB_List=[''.join(choice(Line_of_Business) for _ in range(1)) for _ in range(100)]
BU_List=[''.join(choice(BU_List) for _ in range(1)) for _ in range(100)]
App=pd.DataFrame()
App['LOB']=LOB_List
App['Asset_Type']=Asset_List
App['App State']=Object_List
App['App Name']=App_Name
App['BU']=BU_List
```

```python
####Server Table
chars = ascii_lowercase + digits
seed(20)
status=['Active','Planned','Retired','Active','Active']
system_environment=['Backup Server','Baremetal','Operating System','Installed Software','Storage Array']
server_list = [''.join(choice(chars) for _ in range(8)) for _ in range(500)]
status_list = [''.join(choice(status) for _ in range(1)) for _ in range(500)]
system_list = [''.join(choice(system_environment) for _ in range(1)) for _ in range(500)]
app_id= [''.join(choice(App_Name) for _ in range(1)) for _ in range(500)]
Server_0=pd.DataFrame()
Server_0['Server Name']=server_list
Server_0['Server Status']=status_list
Server_0['System Environment']=system_list
Server_0['App ID']=app_id
Server_0.to_csv('ServerBU.csv')
App.to_csv('Applications.csv')
```
